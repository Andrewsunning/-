```python
import torch
import torch.nn as nn
import torch.utils.data as Data
import torchvision
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
train_data = pd.read_csv('./data/000651.格力电器1 - 副本.csv')
train_data.head()
```

```python
train_data.shape
```

```python
features = [x for x in train_data.columns if x not in ['year', 'month', 'day', 'week']]
features
```

```python
raw_data = train_data[features]
print(raw_data.shape)
raw_data.head()
```

```python
raw_data['zhangdiefu'] = raw_data['zhangdiefu'].replace('#VALUE!', 1)
raw_data['zhangdiefu'] = raw_data.zhangdiefu.astype('float')
raw_data.head()
```

```python
train_x = raw_data[[x for x in raw_data.columns if x != 'Label']]
train_y = raw_data['Label']

train_x.head()
```

```python
# 做数据集标准化

# 初始化缩放器
scaler = MinMaxScaler()

# 拟合，计算缩放操作需要的个字段的最大和最小值
scaler.fit(train_x)
minmax_train_x = scaler.transform(train_x)
minmax_train_x[0]
```

```python
# 将ndarray类型数据转化为tensor类型
train_x = torch.from_numpy(minmax_train_x)
train_y = torch.from_numpy(train_y.values)

# 将DoubleTensor转化为FloatTensor
train_x = train_x.type(torch.FloatTensor)
train_y = train_y.type(torch.FloatTensor)
```

```python
# 设置pytorch的随机种子，保证结果能够复现
torch.manual_seed(1)    # reproducible

# 设置超参数
EPOCH = 10    # 设置迭代训练的轮数
BATCH_SIZE = 260     # 设置每轮迭代训练的样本数
LR = 0.05         # learning rate

# 构建适用于pytorch的数据集
torch_dataset = Data.TensorDataset(train_x, train_y)

# 封装数据集，并对数据集分batch
train_loader = Data.DataLoader(
    dataset=torch_dataset,      # torch TensorDataset format
    batch_size=BATCH_SIZE,      # mini batch size
    shuffle=True,               # random shuffle for training
    num_workers=2,              # subprocesses for loading data
)
```

```python
# 定义自编码器类
class AutoEncoder(nn.Module):
    # 设置网络结构（编码器+解码器）
    def __init__(self):
        super(AutoEncoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Linear(32, 20),
            nn.Tanh(),
            nn.Linear(20, 10),
            nn.Tanh(),
            nn.Linear(10, 8),
#             nn.Tanh(),
#             nn.Linear(12, 3),   # compress to 3 features which can be visualized in plt
        )
        self.decoder = nn.Sequential(
            nn.Linear(8, 10),
            nn.Tanh(),
            nn.Linear(10, 20),
            nn.Tanh(),
            nn.Linear(20, 32),
#             nn.Tanh(),
#             nn.Linear(128, 28*28),
#             nn.Sigmoid(),       # compress to a range (0, 1)
        )
    
    # 设置网络前向传播过程
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return encoded, decoded

# 初始化自编码器
autoencoder = AutoEncoder()
print(autoencoder)

# 设置神经网络的优化函数
optimizer = torch.optim.Adam(autoencoder.parameters(), lr=LR)
# 设置神经网络的损失函数
loss_func = nn.MSELoss()

# 迭代训练
for epoch in range(EPOCH):
    # 每轮需要训练所有的batch数据块
    for step, (x, b_label) in enumerate(train_loader):
        # 将数据块转化为32列
        b_x = x.view(-1, 32)
        b_y = x.view(-1, 32)
        
        # 将一个batch的数据块输入网络中
        encoded, decoded = autoencoder(b_x)
        
        # 计算真实值与解码值的损失
        loss = loss_func(decoded, b_y)      # mean square error
        
        # 网络梯度清零
        optimizer.zero_grad()               # clear gradients for this training step
        
        # 网络反向传播
        loss.backward()                     # backpropagation, compute gradients
        
        # 更新网络参数
        optimizer.step()                    # apply gradients
        
        # 设置没10个batch输出网络的损失
        if step % 10 == 0:
            print('Epoch: ', epoch, '| train loss: %.4f' % loss.data.numpy())
```

```python
# 利用训练好的自编码器对训练集train_x进行降维
encode_train_x,_ = autoencoder(train_x)
encode_train_x.size()
```

```python
# 将降维后的tensor转化为numpy,在转化为DataFrame，最后导出为csv文件
pd.DataFrame(encode_train_x.data.numpy()).to_csv('降维特征.csv', index=None)
```
